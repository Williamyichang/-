```python
import numpy as np 
#np.random.seed(10)
import os
import glob
from PIL import Image
from sklearn.model_selection import train_test_split
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D, ZeroPadding2D
from keras.preprocessing.image import ImageDataGenerator,  array_to_img, img_to_array, load_img
from keras import applications
from keras import optimizers
from sklearn.metrics import classification_report
from keras.regularizers import l2 # L2-regularisation
from keras.layers import BatchNormalization
from keras.callbacks import ModelCheckpoint
from tkinter import *
import tkinter as tk
from pytube import *
import tkinter
import cv2
from matplotlib import pyplot as plt
import pandas as pd
# plot inline
%matplotlib inline

#data_set = "D:/project_data/data_set" # dataset的完整路徑名稱
data_set = "D:/project_data/data_set_1"
top_model_weights_path = 'D:/project_data/temp/weights-improvement-41-0.75-0.821.hdf5'
#img_width, img_height, img_depth = 224,224,3  #告知輸入的image size 
imgSize = (224,224)

dict_labels = {"NG": 0 , "OK": 1} #對應的labels

#num_epochs = 50
#batch_size = 8
l2_lambda = 0.0001
img = [] #存放讀入的圖片資訊
labels = []  #存放原始的labels
label_one = [] #存放轉換為one-hot encoder的labels

#def show_train_history(train_history, train, validation):
    #plt.plot(train_history.history[train])
    #plt.plot(train_history.history[validation])
    #plt.title('Train history')
    #plt.ylabel(train)
    #plt.xlabel('Epoch')
    #legendLoc = 'lower right' if(train=='acc') else 'upper right'
    #plt.legend(['train', 'validation'], loc=legendLoc)
    #plt.show()

def load_image(folder):
    global img, labels, label_one, dict_labels
    for filename in os.listdir(folder):
        label = os.path.basename(folders)
        classname = np.asarray(label)
        if label is not None:
            labels.append(classname)
            label_one.append(dict_labels[label])
        imgs = cv2.imread(os.path.join(folder, filename))
        if imgs is not None:
            img.append(np.array(imgs))

#載入data_set到memory

for folders in glob.glob(data_set+"/*"):
    print(folders)
    load_image(folders)
    
#Step 6.將影像資料及labels轉為Numpy格式
#files = glob.glob (data_set+"/*")

#for myFile in enumerate(files):
#for myFile in files:
    #image = Image.open(myFile)
    #image = image.resize((224, 224), Image.ANTIALIAS)
img = np.array(img)
#img = np.array(img)
print(img.shape[0])
#img = np.resize(img.shape[0],(img.shape[0],img_width, img_height, img_depth)).astype('float32')
#img = np.resize(img.shape[0],img_width, img_height, img_depth),Image.ANTIALIAS.astype('float32')
label_one = np.array(label_one)
print("img.shape={}, label_one.shape =={}".format(img.shape, label_one.shape))

#pic_ID = 60
#plt.imshow(img[pic_ID])
#print("Label:{} , ID:{}, shape:{}".format(labels[pic_ID], label_one[pic_ID], img[pic_ID].shape))

(trainData, testData, trainLabels, testLabels) = train_test_split(img, label_one, test_size=0.2, random_state=42)

print("trainData records: {}".format(len(trainData)))

print("testData records: {}".format(len(testData)))


print("trainData.shape={}  trainLabels.shape={}".format(trainData.shape,trainLabels.shape))

print("testData.shape={}  testLabels.shape={}".format(testData.shape,testLabels.shape))

#針對labels進行獨熱編碼one-hot encoder，並顯示train dataset前五筆編碼後的labels

trainLabels_hot = np_utils.to_categorical(trainLabels)

testLabels_hot = np_utils.to_categorical(testLabels)

print(trainLabels_hot[:5])

#標準化資料：由於圖片點的值介於0~255之間，因此可除以255進行標準化。

trainData_normalize = trainData.astype('float32') / 255.0

testData_normalize = testData.astype('float32') / 255.0

trainData_normalize[0][0][0]

model = Sequential()

model.add(Conv2D(filters=32, kernel_size=(3, 3), padding='same',kernel_regularizer=l2(l2_lambda), input_shape=(imgSize[0], imgSize[1], 3), activation='relu'))
#model.add(Dropout(rate=0.7))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=32, kernel_size=(3, 3), padding='same',kernel_regularizer=l2(l2_lambda), activation='relu'))
#model.add(Dropout(rate=0.7))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=32, kernel_size=(3, 3), padding='same',kernel_regularizer=l2(l2_lambda), activation='relu'))
#model.add(Dropout(rate=0.7))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=32, kernel_size=(3, 3), padding='same',kernel_regularizer=l2(l2_lambda), activation='relu'))
#model.add(Dropout(rate=0.7))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=32, kernel_size=(3, 3), padding='same',kernel_regularizer=l2(l2_lambda), activation='relu'))
#model.add(Dropout(rate=0.7))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

#model.add(Conv2D(filters=64, kernel_size=(3, 3), padding='same', activation='relu'))
#model.add(Dropout(rate=0.7))
#model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dropout(rate=0.7))
model.add(BatchNormalization())
model.add(Dense(256,kernel_regularizer=l2(l2_lambda), activation='relu'))

model.add(Dropout(rate=0.7))

#model.add(Dense(len(dict_labels), activation='softmax'))
#model.add(Dense(1), activation='sigmoid')
model.add(Dense(2))
model.add(Activation('softmax'))

#model.load_weights(model_weights_path)
#datagen = ImageDataGenerator(
        #width_shift_range=0.1,  # randomly shift images horizontally (fraction of total width)
        #height_shift_range=0.1)  # randomly shift images vertically (fraction of total height)
#datagen.fit(trainData_normalize)

model.summary()

#trainParas= optimizers.SGD(lr=0.001, decay=1e-6, momentum=0.9, nesterov=True)

#model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
#model.compile(loss='categorical_crossentropy', optimizer=trainParas, metrics=['accuracy'])

filepath="D:/project_data/temp/weights-improvement-{epoch:02d}-{val_acc:.2f}-{val_loss:.3f}.hdf5"
checkpointer = ModelCheckpoint(filepath,  monitor='val_acc', verbose=1, save_best_only=True, mode='max', period=1)

train_history=model.fit(x=trainData_normalize, y=trainLabels_hot, validation_data=(testData_normalize, testLabels_hot), 
                        validation_split=0.2, epochs=1, batch_size=8, verbose=1,callbacks=[checkpointer])

#train_history = model.fit_generator(datagen.flow(trainData_normalize, trainLabels_hot,
                        #batch_size=batch_size),
                        #steps_per_epoch=trainData_normalize.shape[0]//batch_size,
                        #epochs=num_epochs,
                        #validation_data=(testData_normalize, testLabels_hot),
                        #verbose=1)

def show_train_history(train_history, train, validation):
    plt.plot(train_history.history[train])
    plt.plot(train_history.history[validation])
    plt.title('Train history')
    plt.ylabel(train)
    plt.xlabel('Epoch')
    legendLoc = 'lower right' if(train=='acc') else 'upper right'
    plt.legend(['train', 'validation'], loc=legendLoc)
    plt.show()
    
show_train_history(train_history, 'acc', 'val_acc')
show_train_history(train_history, 'loss', 'val_loss')

prediction = model.predict_classes(testData_normalize)
pd.crosstab(testLabels, prediction, rownames=['label'], colnames=['predict'])
#混淆矩陣
print(pd.crosstab(testLabels, prediction, rownames=['label'], colnames=['predict']))
#classification_report表
print(classification_report(testLabels, prediction))

```
