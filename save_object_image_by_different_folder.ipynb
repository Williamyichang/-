```python
# xml to txt and save the image to different folder for different class 

import os 
import glob
import cv2
import pandas as pd
import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
import numpy as np
%matplotlib
os.getcwd()

def convert_annotation(image_add):
    # image_add进来的是带地址的.jpg
    #image_add = os.path.split(image_add,' ')[1]  # to get file name
    image_name = image_add.split()[0] #to get file name with jpg
    #print(image_name)
    image_name = image_name.replace('.jpg', '')  # replace .jpg by ''
    in_file = open('/media/william/HDD_1TB/experiment/darknet/label_william/anno/' + image_name + '.xml')  # image name vs path
    #out_file = open('/media/william/HDD_1TB/experiment/darknet/label_william/labels/%s.jpg' % (image_name), 'w') # need to put with image in the same folder
 
    tree = ET.parse(in_file)
    root = tree.getroot()
 
    size = root.find('size')
 
    
 
    # To find the class name under object
    for obj in root.iter('object'):
        # To find the class 
        cls = obj.find('name').text
        #cls
        if cls =='person':
            image_dir='/media/william/HDD_1TB/experiment/darknet/label_william/image_1/'
            output_dir = '/media/william/HDD_1TB/experiment/darknet/label_william/test_image/'
            file_name=root.find('filename').text
            image_path = os.path.join(image_dir,file_name)
            print(image_path)
            #for image in image_path:
            im = cv2.imread(image_path)
            #print(im)
            plt.imshow(im)
            plt.savefig('/media/william/HDD_1TB/experiment/darknet/label_william/test_image/%s' % (file_name))
    
 
 
# if not os.path.exists('/media/william/HDD_1TB/experiment/darknet/label_william/test_mage/'):#if no floder
#     os.makedirs('/media/william/HDD_1TB/experiment/darknet/label_william/test_image/') # create labels folder
    
image_adds = open("/media/william/HDD_1TB/experiment/darknet/label_william/train.txt") 
for image_add in image_adds:
    image_add = image_add.strip()
    convert_annotation(image_add)
print("Finished")
```
